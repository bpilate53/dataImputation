```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import scipy as sp
import sklearn as skl
 
import tensorflow as tf
```

```python
full_data = pd.read_csv('Datasets/Real-Time_Road_Conditions.csv')
null_data = pd.read_csv('data_selected_indexes_in_5_columns_0.2_rows_0.5_0.20371799465636653')
```

```python
percentage_of_rows_to_select = 0.5
total_rows_to_select = int(np.floor(full_data.count()[0]*percentage_of_rows_to_select))
full_data = full_data[:total_rows_to_select]
```

```python
null_data.drop(columns=['Unnamed: 0'], inplace=True)
# null_data.drop(columns=['timestamp'], inplace=True)
null_data.head()
```

```python
full_columns = full_data.columns
null_columns = null_data.columns
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# manual one hot encoding

# label_encoder = LabelEncoder()
# new_target = label_encoder.fit_transform(full_data['Location Name'])
# OneHotEncoder().fit(new_target.reshape(-1,1))

# OneHotEncoder().fit_transform(new_target.reshape(-1,1)).toarray()
```

```python
# full data with null datum in selected columns
for column in full_columns:
    if column in null_columns:
        full_data[column] = null_data[column]
```

```python
# full_data.drop(columns=['timestamp'], inplace=True)
## removing id, sensor_id, location 
full_data.drop(columns=['id', 'sensor_id', 'Location', 'timestamp'], inplace=True)
```
