# Real Time Road Conditions Dataset
Austin Transportation road condition sensors 
https://catalog.data.gov/dataset/real-time-road-conditions

```python
#Read real time road conditions dataset
data_set1 = pd.read_csv("Real-Time_Road_Conditions.csv")

#Return last 5 rows of the dataset
data_set1.tail()
```

#### Check for any Null Values

```python
data_set1.isnull().sum()
```

#### Convert Categorical Data to Numeric

```python
from sklearn.preprocessing import LabelEncoder
# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
label1 = le.fit_transform(data_set1['Location Name'])
label2 = le.fit_transform(data_set1['condition_text_displayed'])
label3 = le.fit_transform(data_set1['condition_text_measured'])
label4 = le.fit_transform(data_set1['grip_text'])
label5 = le.fit_transform(data_set1['Location'])


# removing the columns from datasets
# as it is of no use now.
data_set1.drop("Location Name", axis=1, inplace=True)
data_set1.drop("condition_text_displayed", axis=1, inplace=True)
data_set1.drop("condition_text_measured", axis=1, inplace=True)
data_set1.drop("grip_text", axis=1, inplace=True)
data_set1.drop("Location", axis=1, inplace=True)
data_set1.drop("timestamp", axis=1, inplace=True)

# Appending the array to our dataFrame
# with column name
data_set1["Location Name"] = label1
data_set1["condition_text_displayed"] = label2
data_set1["condition_text_measured"] = label3
data_set1["grip_text"] = label4
data_set1["Location"] = label5
data_set1.head()
```

### Select Columns from list and remove cases from the datafame

```python
cols_list = data_set1.columns
limited_cols_list = cols_list[3:8]
limited_cols_list
```

```python
# randomly remove cases from the dataframe
for col in data_set1[limited_cols_list]:
    data_set1.loc[data_set1.sample(frac=0.1).index, col] = np.nan
```

```python
data_set1.head()
```

#### Impute Missing Values Using Extra Trees Regresssor

```python
%%time
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import ExtraTreesRegressor

estimator_rf = ExtraTreesRegressor(n_estimators=10, random_state=0)
#start time
# execution time
x_rf_imputed = IterativeImputer(estimator=estimator_rf, random_state=0, max_iter=50).fit_transform(data_set1)
```

#### Impute Missing Values Using AdaBoostRegressor

```python
%%time
from sklearn.ensemble import AdaBoostRegressor
regr = AdaBoostRegressor(random_state=0, n_estimators=100)
regr_imputed = IterativeImputer(estimator=regr, random_state=0).fit_transform(data_set1)
```
